```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer = sklearn.datasets.load_breast_cancer()
```

```python
breast_cancer_data = pd.DataFrame(breast_cancer.data, columns = breast_cancer.feature_names)
```

```python
breast_cancer_data.head()
```

```python
df=breast_cancer_data.copy()
```

```python
df["label"]=breast_cancer.target
```

```python
x=df.drop("label",axis=1)
y=df["label"]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras

```

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(30,activation='sigmoid'),
    keras.layers.Dense(2,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
train=model.fit(x_train_scaled,y_train,validation_split=0.1,epochs=10,batch_size=32)
```

```python
plt.plot(train.history['accuracy'])
plt.plot(train.history['val_accuracy'])
plt.title('model accuracy')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.legend(['train','validation data'],loc='lower right')
```

```python
loss,accuracy=model.evaluate(x_test_scaled,y_test)

print(accuracy)
```

```python
y_predict=model.predict(x_test_scaled)
```

```python
print(y_predict[0])
```

```python
y_predict_labels=[np.argmax(i) for i in y_predict]
print(y_predict_labels[0])
```
